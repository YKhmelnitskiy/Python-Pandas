```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()

```

## Player Count

```python

total_players = purchase_data["SN"].nunique()

Total_Players = {"Total Players": [total_players]}

TotalPlayers = pd.DataFrame(Total_Players)

TotalPlayers
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Run basic calculations to obtain number of unique items, average price, etc.
number_uniq_items = purchase_data["Item Name"].nunique()
avg_purch_price = purchase_data["Price"].mean()
purchase_count = purchase_data["Purchase ID"].count()
tot_rev = purchase_data["Price"].sum()

#Create a summary data frame to hold the results
Purc_Ana_tot = {"Number of Unique Items": [number_uniq_items],
                            "Average Price":[avg_purch_price],
                            "Number of Purchases":[purchase_count],
                            "Total Revenue": [tot_rev]
               }

Purchasing_Analysis_Total = pd.DataFrame(Purc_Ana_tot)

#Optional: give the displayed data cleaner formatting
Purchasing_Analysis_Total["Average Price"] = Purchasing_Analysis_Total["Average Price"].map("${:.2f}".format)
Purchasing_Analysis_Total["Total Revenue"] = Purchasing_Analysis_Total["Total Revenue"].map("${:,.2f}".format)

#Display the summary data frame
Purchasing_Analysis_Total
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Establish bins for ages
bins = [0,9,14,19,24,29,34,39,100]
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Categorize the existing players using the age bins. Hint: use pd.cut()
dropped_dupl_sn["Age Groups"] = pd.cut(dropped_dupl_sn["Age"], bins,labels=group_labels)

#Calculate the numbers and percentages by age group
Age_group_total = dropped_dupl_sn["Age Groups"].count()
Age_group_count = dropped_dupl_sn["Age Groups"].value_counts()
perc_Age_group_count = ((Age_group_count/Age_group_total)*100)

#Create a summary data frame to hold the results
Age_Demo = pd.DataFrame({"Total Count": Age_group_count,
             "Percentage of Players":perc_Age_group_count,
             })

#Optional: round the percentage column to two decimal points
Age_Demo["Percentage of Players"] = Age_Demo["Percentage of Players"].map("{:.2f}".format)
Age_Demo = Age_Demo.sort_index()

#Display Age Demographics Table
Age_Demo   
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Bin the purchase_data data frame by age
purchase_data["Age Groups"] = pd.cut(purchase_data["Age"], bins,labels=group_labels)

Grouped_Age = purchase_data.groupby(['Age Groups'])

#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
Age_purch_counts = Grouped_Age["Purchase ID"].count()
Age_avgPrice = Grouped_Age["Price"].mean()
Age_totPrice = Grouped_Age["Price"].sum()
Age_avgTotperson = (Age_totPrice/Age_purch_counts)

#Create a summary data frame to hold the results
Purch_Analysis_Age = pd.DataFrame({"Purchase Count": Age_purch_counts,
             "Average Purchase Price":Age_avgPrice,
             "Total Purchase Value":Age_totPrice,
             "Avg Total Purchase per Person":Age_avgTotperson})

#Optional: give the displayed data cleaner formatting
Purch_Analysis_Age["Average Purchase Price"] = Purch_Analysis_Age["Average Purchase Price"].map("${:,.2f}".format)
Purch_Analysis_Age["Total Purchase Value"] = Purch_Analysis_Age["Total Purchase Value"].map("${:,.2f}".format)

Purch_Analysis_Age["Avg Total Purchase per Person"] = Purch_Analysis_Age["Avg Total Purchase per Person"].map("${:,.2f}".format)
#Display the summary data frame
Purch_Analysis_Age  # I don't get how they sorted in their example?
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders = purchase_data.groupby(['SN'])

#Run basic calculations to obtain the results in the table below
s_tot_price = spenders["Price"].sum()
s_purch_count = spenders["Purchase ID"].count()
s_avgTotperson = (s_tot_price/s_purch_count)

#Create a summary data frame to hold the results
Spenders = pd.DataFrame({"Purchase Count": s_purch_count,
             "Average Purchase Price":s_avgTotperson,
             "Total Purchase Value":s_tot_price,
             })

#Sort the total purchase value column in descending order
Spenders = Spenders.sort_values(by='Total Purchase Value',ascending=False)

#Optional: give the displayed data cleaner formatting
Spenders["Total Purchase Value"] = Spenders["Total Purchase Value"].map("${:,.2f}".format)
Spenders["Average Purchase Price"] = Spenders["Average Purchase Price"].map("${:,.2f}".format)

#Display a preview of the summary data frame
Spenders.head(5)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_data = purchase_data.loc[:,['Item ID', 'Item Name','Price']]
grp_item_data = item_data.groupby(['Item ID', 'Item Name'])
Item_Price = grp_item_data["Price"].unique()
Purchase_Count = grp_item_data["Item ID"].count()
Tot_purch_va = grp_item_data["Price"].sum()

popular_items = pd.DataFrame({"Purchase Count": Purchase_Count,
                              "Item Price":Item_Price,
                              "Total Purchase Value": Tot_purch_va
                                })

#Sort the above table by total purchase value in descending order
popular_items = popular_items.sort_values(by='Total Purchase Value',ascending=False)

#Optional: give the displayed data cleaner formatting
popular_items["Total Purchase Value"] = popular_items["Total Purchase Value"].map("${:,.2f}".format)

#Display a preview of the data frame
popular_items.head(5)
```
